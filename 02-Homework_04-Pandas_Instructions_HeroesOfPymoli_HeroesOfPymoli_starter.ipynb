```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/HOP_purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

# Displaying Data Frame 
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data['SN'].unique())
show_player_count = pd.DataFrame({'Player Count': [player_count]})
show_player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Finding/Assigning Data

unique_items = len(purchase_data['Item Name'].unique())
avg_item_price = purchase_data['Price'].mean()
number_of_purchase = len(purchase_data['Item Name'])
revenue = purchase_data['Price'].sum()

# DataFrame for Assigments
purchasing_analysis = pd.DataFrame({'Number of Unique Items': [unique_items], 'Average Price': [avg_item_price], 'Number of Purchases':[number_of_purchase], 'Total Revenue': [revenue]})
purchasing_analysis

purchasing_analysis["Average Price"] = purchasing_analysis["Average Price"].map("${:.2f}".format)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].map("${:.2f}".format)
purchasing_analysis = purchasing_analysis[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]

purchasing_analysis.set_index('Number of Unique Items', inplace=True)
print(purchasing_analysis)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Finding/Assigning Data

duplicate_sn = purchase_data.drop_duplicates(subset='SN', keep='first')
total_players = duplicate_sn['Gender'].count()
male_count = duplicate_sn['Gender'].value_counts()['Male']
fem_count = duplicate_sn['Gender'].value_counts()['Female']
unspecified = total_players - male_count - fem_count

m_percent = (male_count / total_players) * 100
f_percent = (fem_count / total_players) * 100
u_percent = (unspecified / total_players) * 100

# DataFrame for Gender Data

gender_df = pd.DataFrame({"":['Male', 'Female', 'Other/Non-Disclosed'], 'Percentage of Players': [m_percent, f_percent, u_percent], 'Total Count': [male_count, fem_count, unspecified]})
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Gender Grouping

group_df = purchase_data.groupby(['Gender'])

# Finding/Assigning Data

purchase_count = group_df['SN'].count()
avg_purchase = group_df['Price'].mean()
purchase_value = group_df['Price'].sum()

duplicate = purchase_data.drop_duplicates(subset='SN', keep='first')
group_wo_dup = duplicate.groupby(['Gender'])

purchasing = (group_df['Price'].sum() / group_wo_dup['SN'].count())

# DataFrame for new info

purchasing_df = pd.DataFrame({'Purchase Count': purchase_count, 'Average Purchase Price': avg_purchase, 'Total Purchase Value': purchase_value, 'Avg Total Purchase per Person': purchasing})

purchasing_df["Average Purchase Price"] = purchasing_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_df["Avg Total Purchase per Person"] = purchasing_df["Avg Total Purchase per Person"].map("${:.2f}".format)

purchasing_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establishing Bins 

bins = [0,10,15,20,25,30,35,40,100]
bin_cats = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

bin_df = purchase_data.copy()
bin_df['Age Groups'] = pd.cut(bin_df['Age'], bins, labels=bin_cats)
bin_column = pd.cut(bin_df['Age'], bins, labels=bin_cats)
bin_group = bin_df.groupby(['Age Groups'])

# Assign Data

bin_age_counts = bin_group['Age'].count()
bin_player_percentage = (bin_age_counts / total_players) * 100

# Age Demographics DataFrame

age_demographics = pd.DataFrame({'Total Count': bin_age_counts, 'Percentage of Players': bin_player_percentage})
age_demographics['Percentage of Players'] = age_demographics['Percentage of Players'].map("{:.2f}%".format)
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchasing_df = purchase_data
age_purchasing_df['Age Groups'] = pd.cut(purchase_data['Age'], bins, labels = bin_cats)

age_purchasing_analysis_df = age_purchasing_df[['Age', 'Age Groups']].groupby(['Age Groups']).count()
age_purchasing_analysis_df['Average Purchase Price'] = age_purchasing_df[['Price', 'Age Groups']].groupby(['Age Groups']).mean()
age_purchasing_analysis_df['Total Purchase Price'] = age_purchasing_df[['Price', 'Age Groups']].groupby(['Age Groups']).sum()

age_purchasing_analysis_df['Average Purchase Price'] = age_purchasing_analysis_df['Average Purchase Price'].map('${:,.2f}'.format)
age_purchasing_analysis_df['Total Purchase Price'] = age_purchasing_analysis_df['Total Purchase Price'].map('${:,.2f}'.format)

age_purchasing_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Define Top Spender
top_spender = purchase_data['Item ID'].groupby(purchase_data['SN']).count()

# DataFrame 
top_spender= pd.DataFrame(data=top_spender)
top_spender.columns = ['Purchase Count']

top_spender['Average Purchase Price'] = round(purchase_data['Price'].groupby(purchase_data['SN']).mean(),2)
top_spender['Total Purchase Value'] = purchase_data['Price'].groupby(purchase_data['SN']).sum()


top_spender.sort_values(by=['Total Purchase Value'], ascending=False, inplace=True)

top_spender['Average Purchase Price'] = top_spender['Average Purchase Price'].map('${:,.2f}'.format)
top_spender['Total Purchase Value'] = top_spender['Total Purchase Value'].map('${:,.2f}'.format)

top_spender.head(15)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Find/Assign Data
most_items = purchase_data.groupby(['Item ID', 'Item Name','Price'])['Price'].agg(['count','sum'])
most_items.columns = ['Purchase Count', 'Total Purchase Value']

# Remove Index 
most_items.reset_index(inplace=True)
most_items.set_index(['Item ID','Item Name'] ,inplace=True)

most_items = most_items[['Purchase Count', 'Price', 'Total Purchase Value']]

#Sorting by Purchase Count
most_popular = most_items.sort_values(by='Purchase Count', ascending=False)

#DataFrame Formatting 
most_popular['Price'] = most_popular['Price'].map('${:,.2f}'.format)
most_popular['Total Purchase Value'] = most_popular['Total Purchase Value'].map('${:,.2f}'.format)

most_popular.head(15)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Put Data in Desceding Order
most_profitable_items = most_items.sort_values(by='Total Purchase Value', ascending=False)

# DataFrame 
most_profitable_items['Price'] = most_profitable_items['Price'].map('${:,.2f}'.format)
most_profitable_items['Total Purchase Value'] = most_profitable_items['Total Purchase Value'].map('${:,.2f}'.format)
most_profitable_items.head()
```
